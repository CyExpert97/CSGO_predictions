```python
import pandas as pd
import numpy as np
csgo_data_df = pd.read_csv("cleaned_csgo.csv", index_col=0)
csgo_data_df
```

```python
csgo_data_df["time_left"] = csgo_data_df["time_left"].astype(int)
csgo_data_df.info()
csgo_data_df
```

```python
import plotly.express as px
px.imshow(csgo_data_df.corr())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

X = csgo_data_df.drop(["Terrorists_win"], axis=1).values
y = csgo_data_df["Terrorists_win"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.5)

print("Number of samples in:")
print(f"    Training: {len(y_train)}")
print(f"    Validation: {len(y_validation)}")
print(f"    Testing: {len(y_test)}")

# scaler = StandardScaler()
# scaler.fit(X_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
# X_validation = scaler.transform(X_validation)



clf = LogisticRegression(random_state=0, max_iter=100000).fit(X_train, y_train)
clf.score(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X = csgo_data_df.drop(["Terrorists_win"], axis=1).values
y = csgo_data_df["Terrorists_win"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.5)

print("Number of samples in:")
print(f"    Training: {len(y_train)}")
print(f"    Validation: {len(y_validation)}")
print(f"    Testing: {len(y_test)}")


neigh = KNeighborsClassifier(n_neighbors=4)
neigh.fit(X_train, y_train)



y_pred = neigh.predict(X_validation)

from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_validation, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

X = csgo_data_df.drop(["Terrorists_win"], axis=1).values
y = csgo_data_df["Terrorists_win"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.5)

print("Number of samples in:")
print(f"    Training: {len(y_train)}")
print(f"    Validation: {len(y_validation)}")
print(f"    Testing: {len(y_test)}")

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)

# For Decision Trees the result was around 80%
```

```python
csgo_data_df.shape
```

```python
pred_train = clf.predict_proba(X_train)
pred_val = clf.predict_proba(X_validation)

from sklearn.metrics import roc_curve, auc
train_fpr, train_tpr, train_thresholds = roc_curve(y_train, pred_train[:, 1])
val_fpr, val_tpr, val_thresholds = roc_curve(y_validation, pred_val[:, 1])
auc(train_fpr, train_tpr), auc(val_fpr, val_tpr)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'max_depth':[2, 5, 7, 11, 14] , 'min_samples_leaf':[2, 5, 10, 20, 50, 100]}
model_1 = GridSearchCV(clf, parameters, scoring='roc_auc', cv=10)
model_1.fit(X_train, y_train)
```

```python
pred_train = model_1.predict_proba(X_train)
pred_val = model_1.predict_proba(X_validation)

from sklearn.metrics import roc_curve, auc
train_fpr, train_tpr, train_thresholds = roc_curve(y_train, pred_train[:, 1])
val_fpr, val_tpr, val_thresholds = roc_curve(y_validation, pred_val[:, 1])
auc(train_fpr, train_tpr), auc(val_fpr, val_tpr)
```

```python
 prevelence = y_train.sum()/y_train.shape[0]
 cost_fp = 2
 cost_fn = 1
 m = ((1 - prevelence)/prevelence)*(cost_fp/cost_fn)
 m
```

```python
fm = val_tpr - m*val_fpr
fm
```

```python
fm_df = pd.DataFrame(fm)
fm_df['thresholds'] = val_thresholds
fm_df["fpr"] = val_fpr
fm_df["tpr"] = val_tpr
fm_df = fm_df.sort_values(0, ascending=False).reset_index()
```

```python
threshold = fm_df['thresholds'][0]
val_pred = np.where(model_1.predict_proba(X_validation)[:,1]>threshold,1,0)
```

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_validation, val_pred).ravel()
(tn, fp, fn, tp)
```
